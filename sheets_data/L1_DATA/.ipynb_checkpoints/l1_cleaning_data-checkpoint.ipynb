```python
l1_df_2223 = pd.read_csv("/Users/lewisclark/Desktop/PROJECTS/final_project/sheets_data/L1_DATA/ligue_1_data - 22_23.csv", header=1)

l1_df_2223.reset_index(drop=True, inplace=True)

l1_df_2223.head(3)
```

```python
l1_df_2122 = pd.read_csv("/Users/lewisclark/Desktop/PROJECTS/final_project/sheets_data/L1_DATA/ligue_1_data - 21_22.csv", header=1)

l1_df_2122.reset_index(drop=True, inplace=True)

l1_df_2122.head(3)
```

```python
l1_df_2021 = pd.read_csv("/Users/lewisclark/Desktop/PROJECTS/final_project/sheets_data/L1_DATA/ligue_1_data - 20_21.csv", header=1)

l1_df_2021.reset_index(drop=True, inplace=True)

l1_df_2021.head(3)
```

```python
dfs = [l1_df_2425, l1_df_2324, l1_df_2223, l1_df_2122, l1_df_2021]

keep_columns = [
    'Squad', 'Poss', '90s', 'Gls.1', 'Ast.1', 'G+A.1', 'G-PK.1',
    'xG.1', 'xAG.1', 'xG+xAG', 'npxG+xAG.1', 'PrgC', 'PrgP', 'CrdY', 'CrdR']

l1_df_2425, l1_df_2324, l1_df_2223, l1_df_2122, l1_df_2021 = [df[keep_columns] for df in dfs]

l1_df_2425.head()
```

```python
l1_df_2425['Season'] = '2024-25'
l1_df_2324['Season'] = '2023-24'
l1_df_2223['Season'] = '2022-23'
l1_df_2122['Season'] = '2021-22'
l1_df_2021['Season'] = '2020-21'

master_l1_df_stats = pd.concat([l1_df_2425, l1_df_2324, l1_df_2223, l1_df_2122, l1_df_2021], ignore_index=True)
```
