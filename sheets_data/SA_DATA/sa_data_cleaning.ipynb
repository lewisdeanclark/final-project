```python
dfs = [sa_df_2425, sa_df_2324, sa_df_2223, sa_df_2122, sa_df_2021]

keep_columns = [
    'Squad', 'Poss', '90s', 'Gls.1', 'Ast.1', 'G+A.1', 'G-PK.1',
    'xG.1', 'xAG.1', 'xG+xAG', 'npxG+xAG.1', 'PrgC', 'PrgP', 'CrdY', 'CrdR']

sa_df_2425, sa_df_2324, sa_df_2223, sa_df_2122, sa_df_2021 = [df[keep_columns] for df in dfs]

sa_df_2425.head()
```

```python
sa_df_2425['Season'] = '2024-25'
sa_df_2324['Season'] = '2023-24'
sa_df_2223['Season'] = '2022-23'
sa_df_2122['Season'] = '2021-22'
sa_df_2021['Season'] = '2020-21'

master_sa_df_stats = pd.concat([sa_df_2425, sa_df_2324, sa_df_2223, sa_df_2122, sa_df_2021], ignore_index=True)
```
